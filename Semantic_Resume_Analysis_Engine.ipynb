````python
import google.generativeai as genai
import json
from google.colab import userdata # Or use os.getenv for local scripts
import PyPDF2

# --- CONFIGURATION ---
try:
    # Use Colab secrets to securely store your API key
    MY_API_KEY = userdata.get('GEMINI_API_KEY')
    genai.configure(api_key=MY_API_KEY)
    print("API key configured.")
except Exception as e:
    print(f"Error configuring API key. Please set it up in Colab secrets. {e}")


# --- INPUTS ---
# Ask the user for the job description text first
job_description_text = input("Please paste the job description here: ")

# Now, ask the user to upload their resume
print("\nPlease upload your resume file.")
from google.colab import files
uploaded = files.upload()

# Read the content of the uploaded PDF file
uploaded_filename = next(iter(uploaded)) # Assuming only one file was uploaded
with open(uploaded_filename, 'rb') as f:
    reader = PyPDF2.PdfReader(f)
    original_resume_text = ""
    for page_num in range(len(reader.pages)):
        original_resume_text += reader.pages[page_num].extract_text()


# --- THE ULTIMATE PROMPT ---
# This prompt asks the model to perform all tasks: score, analyze, and rewrite.
prompt = f"""
You are an expert AI resume writer and career coach. Your task is to analyze the provided "Original Resume" against the "Job Description". Then, rewrite the resume to perfectly match the tone, keywords, and qualifications of the job description.

Produce a JSON object with the following four keys:
1.  "match_score": An integer score from 0 to 100 representing how well the original resume matches the job description.
2.  "analysis_summary": A brief summary of the original resume's strengths and weaknesses for this specific role.
3.  "key_changes_made": A bulleted list (as a single string) detailing the specific changes you made and why, referencing keywords from the job description.
4.  "optimized_resume": The full text of the newly rewritten and optimized resume. Use markdown for formatting (e.g., bolding for titles, bullet points for experience).

---
**JOB DESCRIPTION:**
{job_description_text}
---
**ORIGINAL RESUME:**
{original_resume_text}
---
"""


# --- API CALL AND OUTPUT ---
print("\nü§ñ Running full analysis and optimization...")
try:
    model = genai.GenerativeModel('gemini-2.5-flash')
    response = model.generate_content(prompt)

    # Clean up the response to make it valid JSON
    cleaned_json_string = response.text.strip().replace("```json", "").replace("```", "")

    # Parse the JSON string into a Python dictionary
    analysis = json.loads(cleaned_json_string)

    print("\n" + "="*40)
    print("‚úÖ Co-Pilot Analysis Complete")
    print("="*40)

    print(f"\n## üìä Original Match Score: {analysis['match_score']}/100")

    print("\n## üìù Analysis Summary")
    print(analysis['analysis_summary'])

    print("\n## üîß Key Changes Made to Optimize")
    print(analysis['key_changes_made'])

    print("\n" + "="*40)
    print("üìÑ Your New, Optimized Resume")
    print("="*40)
    print(analysis['optimized_resume'])


except Exception as e:
    print(f"An error occurred: {e}")
````
